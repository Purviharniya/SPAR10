```
# Configure daknet to use GPU and OpenCV 
# all commands starting with '!' are Unix commands. We need to add the symbol while running it in Colab
# sed: string editor, -i: inplace : replaces the string in the given file instead of making a copy 
# so basically in these commands, we search for all the strings where opencv, gpu and cudnn = 0 and replace it with 1
%cd darknet
!sed -i 's/OPENCV=0/OPENCV=1/' Makefile
!sed -i 's/GPU=0/GPU=1/' Makefile
!sed -i 's/CUDNN=0/CUDNN=1/' Makefile
```

```
# Compile the changes
!make
```
